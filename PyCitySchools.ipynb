```python
#student prefixes and sufixes to erase from the name
prefix_sufix = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]

#erasing the list of prefix and sufix out of the names from the original dataframe
for word in prefix_sufix:
    student_data_df["student_name"]=student_data_df["student_name"].str.replace(word,"",regex=True)
    
student_data_df
```

```python
#Getting the new student data out of other file provided in the course
clean_student_data_df = pd.read_csv(clean_student_data)
clean_student_data_df

#Erasing first "index" column of the file
clean_student_data_df=clean_student_data_df.drop(clean_student_data_df.columns[0], axis=1)
clean_student_data_df
```

```python
#Merging the two dataframes into one
school_data_complete = pd.merge(clean_student_data_df, school_data_df, on=["school_name","school_name"])
school_data_complete.head()
```

```python
#Counting the students in the merged dataframe
student_count = school_data_complete.count()
student_count
```

```python
#Choosing one column to count studends due to all columns have the same count
student_count = school_data_complete["Student ID"].count()
student_count
```

```python
#Counting the total schools
school_count_2 = school_data_complete["school_name"].unique()
len(school_count_2)
```

```python
#Summin all the budget from the original school data dataframe
total_budget = school_data_df["budget"].sum()
total_budget
```

```python
# Getting the average score for reading of the complete dataframe
average_reading_score = school_data_complete["reading_score"].mean()
average_reading_score
```

```python
# Getting the average score for math of the complete dataframe
average_math_score = school_data_complete["math_score"].mean()
average_math_score
```

```python
# Assign boolean varialbles to a pasing grade above 70
passing_math = school_data_complete["math_score"] >=70
passing_reading = school_data_complete["reading_score"] >=70
```

```python
#Creating a new dataframe with just the passing scores of math and reading 
passing_math = school_data_complete[school_data_complete["math_score"] >=70]
passing_reading = school_data_complete[school_data_complete["reading_score"] >=70]
```

```python
#Counting all the student names for each dataframe to calculate the number of passing grades of each class
passing_math_count = passing_math["student_name"].count()
passing_reading_count = passing_reading["student_name"].count()
print(passing_math_count,passing_reading_count)
```

```python
# Getting the percentage of students that pass each class
passing_math_percentage = float(passing_math_count)/float(student_count)*100
passing_reading_percentage = float(passing_reading_count)/float(student_count)*100
print(passing_math_percentage)
print(passing_reading_percentage)
```

```python
#Get the overall percentage of passing students
passing_math_reading = school_data_complete[(school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70)]
overall_passing_count = passing_math_reading["student_name"].count()
overall_passing_count
```

```python
#Getting overal percentage of passing math and reading scores
overall_passing_percentage = float(overall_passing_count)/float(student_count)*100
overall_passing_percentage
```

```python
#Creating a new table with the summary data
district_summary_df = pd.DataFrame([{"Total Schools":len(school_count_2), "Total Students":student_count,
                       "Total Budget":float(total_budget),"Average Reading Score":float(average_reading_score), 
                       "Average Math Score":float(average_math_score), "% Passing Reading":float(passing_reading_percentage),
                       "% Passing Math":float(passing_math_percentage),"% Overall Passing":float(overall_passing_percentage)}])
district_summary_df
```

```python
#Changing formats
district_summary_df["Total Students"]=district_summary_df["Total Students"].map("{:,}".format)
district_summary_df
```

```python
#Changing formats
district_summary_df["Total Budget"]=district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df
```

```python
#Changing formats
district_summary_df["Average Reading Score"]=district_summary_df["Average Reading Score"].map("{:.1f}".format)
district_summary_df["Average Math Score"]=district_summary_df["Average Math Score"].map("{:.1f}".format)
district_summary_df
```

```python
#Changing formats
district_summary_df["% Passing Reading"]=district_summary_df["% Passing Reading"].map("{:.0f}".format)
district_summary_df["% Passing Math"]=district_summary_df["% Passing Math"].map("{:.0f}".format)
district_summary_df["% Overall Passing"]=district_summary_df["% Overall Passing"].map("{:.0f}".format)
district_summary_df
```

```python
#Reorder columns
new_order=["Total Schools","Total Students","Total Budget","Average Math Score","Average Reading Score",
          "% Passing Math","% Passing Reading","% Overall Passing"]
district_summary_df = district_summary_df[new_order]
district_summary_df
```
